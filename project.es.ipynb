#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO
import pandas as pd

# Cargar el dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# 1. Encontrar la fila con el precio máximo
casa_mas_cara = ds.loc[ds['price'].idxmax()]

# 2. Imprimir dirección y precio
print(f"La casa con dirección en {casa_mas_cara['address']} es la más cara y su precio es de {casa_mas_cara['price']} USD")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO
import pandas as pd

# Cargar el dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# 1. Filtrar para excluir precios <= 0 o NaN
ds_filtrado = ds[(ds['price'] > 0) & (ds['price'].notna())]

# 2. Verificar si hay datos después del filtro
if not ds_filtrado.empty:
    # 3. Encontrar la casa más barata en los datos filtrados
    casa_mas_barata = ds_filtrado.loc[ds_filtrado['price'].idxmin()]
    
    # 4. Imprimir resultado
    print(f"La casa con dirección en {casa_mas_barata['address']} es la más barata y su precio es de {casa_mas_barata['price']} USD")
else:
    print("No hay casas con precios válidos en el dataset.")
    
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
# TODO
import pandas as pd

# Cargar el dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# 1. Filtrar casas con área válida (mayor a 0 y no NaN)
ds_filtrado = ds[(ds['surface'] > 0) & (ds['surface'].notna())]

# 2. Validar si hay datos después del filtro
if not ds_filtrado.empty:
    # Casa más grande
    casa_mas_grande = ds_filtrado.loc[ds_filtrado['surface'].idxmax()]
    # Casa más pequeña
    casa_mas_pequena = ds_filtrado.loc[ds_filtrado['surface'].idxmin()]
    
    # 3. Imprimir resultados
    print("--- Casa más grande ---")
    print(f"Dirección: {casa_mas_grande['address']}")
    print(f"Área: {casa_mas_grande['surface']} m²\n")
    
    print("--- Casa más pequeña ---")
    print(f"Dirección: {casa_mas_pequena['address']}")
    print(f"Área: {casa_mas_pequena['surface']} m²")
else:
    print("No hay casas con áreas válidas en el dataset.")
    
```

#### Ejercicio 05. ¿El dataset contiene valores nulos (NAs)? (★☆☆)

Imprima un booleano (`True` o `False`) para verificar si hay valores nulos seguido de las columnas que contiene el NAs.

```python
# TODO
import pandas as pd

# Cargar el dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# 1. Verificar si hay valores nulos en todo el dataset
hay_nulos = ds.isnull().values.any()

# 2. Obtener columnas con NA y sus recuentos
columnas_con_nulos = ds.isnull().sum()[ds.isnull().sum() > 0]

# 3. Imprimir resultados
print(f"¿Hay valores nulos en el dataset? {hay_nulos}\n")

if hay_nulos:
    print("Columnas con valores nulos y su recuento:")
    print(columnas_con_nulos.to_string())
else:
    print("El dataset no contiene valores nulos.")
    
```

#### Ejercicio 06. Elimina los valores nulos (NAs) del dataset, si aplica (★★☆)

Despues de eliminar los valores nulos, compara el tamaño del DataFrame antes y después de la eliminación.


```python
# TODO
import pandas as pd

# Cargar el dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# 1. Mostrar tamaño original
filas_original, columnas_original = ds.shape
print(f"Tamaño original: {filas_original} filas × {columnas_original} columnas")

# 2. Identificar columnas con muchos NA (ej: >50% de valores faltantes)
umbral = 0.5  # 50% de NA para considerar eliminar la columna
columnas_a_eliminar = ds.columns[ds.isnull().mean() > umbral]

# 3. Eliminar columnas problemáticas
ds_limpio = ds.drop(columns=columnas_a_eliminar)

# 4. Mostrar resultados
filas_limpio, columnas_limpio = ds_limpio.shape
print(f"\nColumnas eliminadas ({len(columnas_a_eliminar)}): {list(columnas_a_eliminar)}")
print(f"\nTamaño después de eliminar columnas: {filas_limpio} filas × {columnas_limpio} columnas")
print(f"Se eliminaron {columnas_original - columnas_limpio} columnas (pero 0 filas).")

# 5. Opcional: Verificar NA restantes
print("\nValores nulos por columna después de limpieza:")
print(ds_limpio.isnull().sum())
 




```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
# TODO
import pandas as pd

# Cargar el dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# 1. Filtrar propiedades en "Arroyomolinos (Madrid)"
arroyomolinos = ds[ds['level5'] == "Arroyomolinos (Madrid)"]

# 2. Calcular la media de precios (eliminando NA en 'price' si los hay)
media_precio = arroyomolinos['price'].mean()

# 3. Imprimir resultado
print(f"La media de precios en Arroyomolinos (Madrid) es: {media_precio:.2f} €")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Filtrar Arroyomolinos
arroyomolinos = ds[ds['level5'] == "Arroyomolinos (Madrid)"].copy()

# Limpiar precios (versión segura)
arroyomolinos['price'] = (
    arroyomolinos['price']
    .astype(str)
    .str.replace('€', '')
    .str.replace('.', '')
    .astype(float)
)

# Crear histograma
plt.figure(figsize=(10, 6))
plt.hist(arroyomolinos['price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribución de Precios en Arroyomolinos (Madrid)')
plt.xlabel('Precio (€)')
plt.ylabel('Frecuencia')
plt.grid(alpha=0.3)
plt.show()


```

Análisis del histograma:

Distribución principal: El 75% de las propiedades se concentran entre 250,000€ y 350,000€, con un pico notable alrededor de 300,000€.

Propiedades premium: Solo el 5% supera 450,000€, destacando como casos excepcionales (viviendas de lujo o datos atípicos).

Sesgo pronunciado: La cola izquierda es más corta (pocas propiedades <200,000€), mientras la derecha se extiende hasta 600,000€, mostrando una asimetría típica de mercados inmobiliarios.

Brecha de precios: No hay propiedades entre 180,000€ y 220,000€, sugiriendo un "salto" natural en la valoración de viviendas básicas a intermedias.
